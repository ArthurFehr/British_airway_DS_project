```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests

from functions import parse_url_text, value_fetcher, rating_fetcher, review_fetcher
```

```python
global rating, aircraft, seat_type, travel_type, route, date_flown, comfort, staff, food, entertainment, wifi, ground_service, vfm, recommend

verified=[]
rating=[]
feedback=[]
aircraft=[]
seat_type=[]
travel_type=[]
route=[]
date_flown=[]
comfort=[]
staff=[]
food=[]
entertainment=[]
wifi=[]
ground_service=[]
value_for_money=[]
recommend=[]
```

```python
headers_text = ['Aircraft', 'Type Of Traveller', 'Seat Type', 'Route', 'Date Flown', 'Recommended']
fields_text = [aircraft, travel_type, seat_type, route, date_flown, recommend]

headers_star = ['Seat Comfort', 'Cabin Staff Service', 'Food & Beverages','Inflight Entertainment',
                 'Ground Service', 'Value For Money']
fields_star = [comfort, staff, food, entertainment, ground_service, value_for_money]
```

```python
for i in range(1, 361):
    articles = parse_url_text(i)
    for article in articles:
        div = article.find('div', class_ = 'tc_mobile')
        table=div.find('table',class_='review-ratings')
        value_fetcher(table, headers_text, fields_text, headers_star, fields_star)
        verification, review = review_fetcher(div)
        verified.append(verification)
        feedback.append(review)
        rating.append(rating_fetcher(article))
```

```python
keys = ["Verified"] + headers_text + headers_star + ['Rating', 'Review']
values = [verified] + fields_text + fields_star + [rating, feedback]

data = pd.DataFrame(dict(zip(keys, values)))

data.head()
```

```python
data.to_csv('../02_raw_data/British_Airways_review_data.csv',index=False)
```
